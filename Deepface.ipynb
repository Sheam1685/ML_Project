```python
from deepface import DeepFace
import os
from sklearn.metrics import accuracy_score, classification_report
```

```python
data_path = "C:\\Users\\samia\\Documents\\ML_Project\\data"
test_data_path = os.path.join(data_path, "test")
emotions = os.listdir(test_data_path)
```

```python
true_labels = []
predicted_labels = []
```

```python
for emotion in emotions:
    print(f"Processing {emotion} images")
    
    emotion_path = os.path.join(test_data_path, emotion)
    images = [os.path.join(emotion_path, image) for image in os.listdir(emotion_path) if image.endswith(".jpg")]
    true_labels.extend([emotion] * len(images))
    
    
    for image in images:
      prediction = DeepFace.analyze(image, actions=['emotion'], enforce_detection=False)
      predicted_labels.append(prediction[0]['dominant_emotion'])
    
    print(f"Done processing {emotion} images")
    
print("Done processing all images")
```

```python
accuracy = accuracy_score(true_labels, predicted_labels)
classification_report = classification_report(true_labels, predicted_labels)

print(f"Accuracy: {accuracy}")
print(f"Classification Report: {classification_report}")
```
